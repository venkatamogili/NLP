```python
import gensim.downloader as api

model = api.load("word2vec-google-news-300")
```

**Exercise 1: Explore a Word Vector**

```python
vector = model['cat']
print(vector)
print(f"Vector dimension: {len(vector)}")
```

**Check Word Existence**

```python
print('cat' in model)
print('xyz123' in model)
```

**Find Similar Words**

```python
similar = model.most_similar("happy", topn=5)
print(similar)
```

**Word Analogies - Basic**

```python
result = model.most_similar(positive=['woman', 'king'], negative=['man'], topn=1)
print(result)
```

**Calculate Similarity**

```python
sim = model.similarity('cat', 'dog')
print(f"Similarity: {sim}")
```

**Odd One Out**

```python
odd = model.doesnt_match(['breakfast', 'lunch', 'dinner', 'computer'])
print(odd)
```

**Country-Capital Analogies**

```python
result = model.most_similar(positive=['Paris', 'Germany'], negative=['France'], topn=1)
print(result)
```

**Verb Forms**

```python
result = model.most_similar(positive=['swimming', 'run'], negative=['running'], topn=1)
print(result)
```

**Comparative Adjectives**

```python
result = model.most_similar(positive=['biggest', 'small'], negative=['big'], topn=1)
print(result)
```

**Find Similarities in Groups**

```python
words = ['cat', 'dog', 'lion']
for word in words:
    print(f"\nSimilar to {word}:")
    print(model.most_similar(word, topn=3))
```

**Word Addition**

```python
result = model.most_similar(positive=['king', 'queen'], topn=3)
print(result)
```

**Compare Animal Similarities**

```python
animals = ['cat', 'dog', 'elephant']
for i in range(len(animals)):
    for j in range(i+1, len(animals)):
        sim = model.similarity(animals[i], animals[j])
        print(f"{animals[i]} vs {animals[j]}: {sim:.3f}")
```

**Find Professional Relationships**

```python
result = model.most_similar(positive=['hospital', 'teacher'], negative=['school'], topn=3)
print(result)
```

**Gender Relationships**

```python
result = model.most_similar(positive=['actress', 'man'], negative=['woman'], topn=1)
print(result)
```

**Explore Word Neighborhoods**

```python
word = "python"
if word in model:
    similar = model.most_similar(word, topn=10)
    for w, score in similar:
        print(f"{w}: {score:.3f}")
```
